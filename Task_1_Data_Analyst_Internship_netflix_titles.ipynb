### Identify and Handle Missing Values

```python
print("\nMissing values before handling:")
print(df.isnull().sum())

df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna('Unknown', inplace=True)
df['rating'].fillna('Unknown', inplace=True)
df['duration'].fillna('Unkown', inplace=True)

print("\nMissing values after handling:")
print(df.isnull().sum())
```

```python
df
```

### Remove Duplicate Rows

```python
print(f"\nNumber of rows before dropping duplicates: {len(df)}")

df.drop_duplicates(inplace=True)

print(f"Number of rows after dropping duplicates: {len(df)}")
```

### 3. Standardize Text Values

```python
df['country'] = df['country'].replace({'USA': 'United States', 'U.S.': 'United States'})

print("\nExample of standardized 'country' values:")
print(df['country'].value_counts().head())
```

### Check and Fix Data Types

```python
print("\nData type of 'date_added' before conversion:")
print(df['date_added'].dtype)

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

print("\nData type of 'date_added' after conversion:")
print(df['date_added'].dtype)
```
